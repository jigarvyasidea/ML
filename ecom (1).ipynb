```python
# done to convert into into lovercase 
df['query'] = df['query'].apply(lambda x: str(x) if pd.notnull(x) else '')
```

```python
df['query'] = df['query'].fillna('')
```

```python
df
```

```python
import string

# Function to preprocess text
def preprocess_text(text):
    # Lowercase and remove punctuation
    text = text.lower().translate(str.maketrans('', '', string.punctuation))
    return text

# Apply preprocessing
df['Processed_Query'] = df['query'].apply(preprocess_text)

print(df[['query', 'Processed_Query']].head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Encode the intents (labels)
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Intent'] = label_encoder.fit_transform(df['Intent'])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['Processed_Query'], df['Intent'], test_size=0.2, random_state=42)

# Convert text to TF-IDF features
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Train a Naive Bayes classifier
model = MultinomialNB()
model.fit(X_train_tfidf, y_train)

# Predict on the test set
y_pred = model.predict(X_test_tfidf)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
# Function to predict intent
def predict_intent(query):
    processed_query = preprocess_text(query)
    query_tfidf = vectorizer.transform([processed_query])
    intent_id = model.predict(query_tfidf)[0]
    intent = label_encoder.inverse_transform([intent_id])[0]
    return intent

# Test the model
query = "How want refun on payment??"
predicted_intent = predict_intent(query)
print(f"Predicted Intent: {predicted_intent}")
```
