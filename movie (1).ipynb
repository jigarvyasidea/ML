```python
def recommend(movies):
    # Find the index of the given movie title
    movie_index = df_new[df_new['title'] == movies].index[0]
    
    # Get the similarity scores for that movie
    distance = similarity[movie_index]
    
    # Sort the movies by similarity scores in descending order and exclude the first one (itself)
    movies_list = sorted(list(enumerate(distance)), reverse=True, key=lambda x: x[1])[1:6]
    
    # Print the titles of the recommended movies
    for i in movies_list:
        print(df_new.iloc[i[0]]['title'])  # Fetch the title using the index
```

```python
df_new = recommend('Avatar')
```
