# Elbow method 

```python
from sklearn.cluster import KMeans 
```

```python
#WCSSS ki list bna deta hu 

Wcss = [ ]

for i in range(1,11):
    # kmena ka object banate h 
    km = KMeans(n_clusters=i)
    # train on dataset 
    km.fit_predict(df)
    # add in emppty list me 
    Wcss.append(km.inertia_)
    
```

```python
Wcss
```

```python
plt.plot(range(1,11),Wcss)
```
